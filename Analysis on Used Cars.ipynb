# **6. Data Pre-Processing**

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

```python
# Insert your code here... 
# Is there any null value in data set?
Car_data.isnull().any()
```

```python
# Engine Version plot
Car_data.engV.hist()
```

```python
# Type of drive in Car
Car_data.drive.hist()
```

```python
# Type of car
Car_data.car.hist()
```

```python
# Mileage of the Car
Car_data.mileage.hist()
```

```python
# Is there any null value in engv column
Car_data.engV.isnull().any()
```

```python
# Is there any null value in price column
Car_data.price.isnull().any()
```

```python
# Replace 'Zero' values in the price and mileage column with NaN
Car_data = Car_data.replace({
        'price': 0,
        'mileage': 0,
    }, np.nan)
```

```python
# Fill NaN value with mode value
Car_data['engV']=Car_data['engV'].fillna(Car_data['engV'].mode()[0])
```

```python
# Check the NaN value in engV column
Car_data.engV.isnull().any()
```

```python
# Fill NaN value with mode value
Car_data['drive']=Car_data['drive'].fillna(Car_data['drive'].mode()[0])
```

```python
# Check the NaN value in drive column
Car_data.drive.isnull().any()
```

```python
# Fill NaN value with mode value
Car_data['price']=Car_data['price'].fillna(Car_data['price'].mode()[0])
```

```python
# Fill NaN value with mode value
Car_data['mileage']=Car_data['mileage'].fillna(Car_data['mileage'].mode()[0])
```

```python
# Check the null value in Data set
Car_data.isnull().any()
```

```python
# Drop the duplicate values
Car_data.drop_duplicates(keep = False, inplace = True)
```

```python
# Check the duplicate data
duplicated = Car_data[Car_data.duplicated()]
```

```python
# Print the info of the data 
Car_data.info()
```

```python
# Brief statistics of data set
Car_data.describe()
```

```python
'''
fig = plt.figure(figsize = (10,6))
sns.catplot(x = 'price', data = Car_data, kind='count')'''
```

```python
# Taking out the extreem values(outliers)
min_out_pric, max_out_pric = Car_data.price.quantile([0.05, 0.95])
min_out_pric, max_out_pric
```

```python
# Remove the outliers
Car_data2 = Car_data[(Car_data.price<max_out_pric) & (Car_data.price>min_out_pric)]
```

```python
# Check the statistics of the data 
Car_data2.describe()
```

```python
# Taking out the extreem values(outliers)
min_out_mil, max_out_mil = Car_data2.mileage.quantile([0.05, 0.95])
min_out_mil, max_out_mil
```

```python
# Remove the outliers
Car_data2 = Car_data2[(Car_data2.mileage<max_out_mil) & (Car_data2.mileage>min_out_mil)]
```

```python
# Check the statistics of the data 
Car_data2.describe()
```
